```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, classification_report
import joblib
```

```python
credit_data = pd.read_csv('data/creditcard.csv')
credit_data.head()
```

```python
credit_data['Class'].value_counts()
```

```python
legit = credit_data[credit_data['Class'] == 0]
fraud = credit_data[credit_data['Class'] == 1]
print(legit.shape)
print(fraud.shape)
```

```python
credit_data.groupby('Class').mean()
```

```python
X_full = credit_data.drop('Class', axis=1)
y_full = credit_data['Class']

x_train_full, x_test, y_train_full, y_test = train_test_split(
    X_full, y_full, test_size=0.2, stratify=y_full, random_state=2
)

print(x_train_full.shape)
print(x_test.shape)
```

```python
y_test.value_counts()
```

```python
train_data = pd.concat([x_train_full, y_train_full], axis=1)
legit_train = train_data[train_data['Class'] == 0]
fraud_train = train_data[train_data['Class'] == 1]

legit_train_sample = legit_train.sample(n=len(fraud_train), random_state=2)
balanced_train = pd.concat([legit_train_sample, fraud_train], axis=0)

x_train = balanced_train.drop('Class', axis=1)
y_train = balanced_train['Class']

y_train.value_counts()
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
log_model = LogisticRegression(max_iter=1000)
log_model.fit(x_train_scaled, y_train)
```

```python
log_train_pred = log_model.predict(x_train_scaled)
log_train_accuracy = accuracy_score(y_train, log_train_pred)
print("Logistic Regression Training Accuracy:", log_train_accuracy)
```

```python
log_test_pred = log_model.predict(x_test_scaled)
log_test_accuracy = accuracy_score(y_test, log_test_pred)
print("Logistic Regression Test Accuracy:", log_test_accuracy)
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, log_test_pred))
print("Precision:", precision_score(y_test, log_test_pred))
print("Recall:", recall_score(y_test, log_test_pred))
print("F1 Score:", f1_score(y_test, log_test_pred))
print(classification_report(y_test, log_test_pred))
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=2)
rf_model.fit(x_train_scaled, y_train)
```

```python
rf_test_pred = rf_model.predict(x_test_scaled)
rf_test_accuracy = accuracy_score(y_test, rf_test_pred)
print("Random Forest Test Accuracy:", rf_test_accuracy)
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, rf_test_pred))
print("Precision:", precision_score(y_test, rf_test_pred))
print("Recall:", recall_score(y_test, rf_test_pred))
print("F1 Score:", f1_score(y_test, rf_test_pred))
print(classification_report(y_test, rf_test_pred))
```

```python
best_model = rf_model if f1_score(y_test, rf_test_pred) > f1_score(y_test, log_test_pred) else log_model
joblib.dump(best_model, 'fraud_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
```
